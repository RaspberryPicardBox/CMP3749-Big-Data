```python
import re

txt_file = "drive/MyDrive/Lincoln/Teaching/2021-2022/BigData/1661-0.txt"

with open(txt_file, "r", encoding="utf-8") as f:
    data = f.read()

clean = re.sub(r"[,.;@#?!&$/â€”-]+\ *", " ", data)
words = clean.split()

print(len(words))

print(words[10000:10010])
```

```python
def find_longest_word(strings):

    longest_word = None
    longest_length = 0

    for s in strings:
        if len(s) > longest_length:
            longest_length = len(s)
            longest_word = s

    return longest_word
```

```python
%time print(find_longest_word(words))
```

```python
corpus = words * 400
print(len(corpus))
%time print(find_longest_word(corpus))
```

```python
%%time

word_lengths = [len(s) for s in corpus]
word_lengths = zip(corpus, word_lengths)
longest = max(word_lengths, key=lambda w: w[1])
print(longest)
```

```python
from functools import reduce

# Define our two functions for map and reduce
mapper = len

def reducer(a, b):

    if a[1] > b[1]:
        return a
    return b
```

```python
%%time

# Map! 
intermediate = map(mapper, corpus)
intermediate = zip(corpus, intermediate)

# Reduce
reduced = reduce(reducer, intermediate)
print(reduced)
```

```python
from multiprocessing import Pool, cpu_count

def chunkify(lst, n):

    for i in range(0, len(lst), n):
        yield lst[i:i + n]


def chunk_mapper(chunk):

    mapped = map(mapper, chunk)
    mapped = zip(chunk, mapped)
    return reduce(reducer, mapped)

no_chunks = cpu_count()
chunks = list(chunkify(corpus, len(corpus) // no_chunks))
```

```python
%%time

pool = Pool(no_chunks)
mapped = pool.map(chunk_mapper, chunks)
print([m for m in mapped])
reduced = reduce(reducer, mapped)
print(reduced)
```
